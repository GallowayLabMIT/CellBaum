```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# NOTE(arl): This should be from your image segmentation code
objects = ('/Users/ConradOakes/Desktop/Galloway_2021/cell_data/cell_locationsIdentifyPrimaryObjects.csv')
objects = pd.read_csv(objects)
```

```python
objects
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
deci_object = objects[(objects['Metadata_time'] < 10) & (objects['Metadata_zstep'] == 1) & (objects['Location_Center_X'] < 500)]
ax.scatter(deci_object['Location_Center_X'], deci_object['Location_Center_Y'], deci_object['Metadata_time'])
```

```python
for t in np.unique(objects['Metadata_time']): 
    deci_object = objects[(objects['Metadata_time'] == t) & (objects['Metadata_zstep'] == 1)]
    plt.scatter(deci_object['Location_Center_X'], deci_object['Location_Center_Y'])
    plt.title(t)
    plt.show()
```
